```python
file2 = open('names.txt','r')
for i in file2:
    list_school = i
list_school = list_school[1:-1]
list_school = list_school.replace("'", '').split(',')
list_school
```

```python
years = ['2021', '2020', '2019', '2018', '2017', '2017', '2016', '2015', '2014', '2013', '2012', '2011']
```

```python
for school in tqdm(list_school, total=len(list_school)):
    
    for year in years:
        try:
            request_url_log = f'https://www.sports-reference.com/cbb/schools/{school}/{year}-gamelogs.html'
            request_url_advandace_log = f'https://www.sports-reference.com/cbb/schools/{school}/{year}-gamelogs-advanced.html'
            response_log = requests.get(request_url_log, headers=headers)
            page_log = response_log.text
            #print(page_log)
            #print('log')
            f = open(f'HTMLs/{school}_{year}_gamelogs.html', 'w+')
            f.write(page_log)
            f.close()
            response_advandace_log = requests.get(request_url_advandace_log, headers=headers)
            #print(response_advandace_log)
            page_advandace_log = response_advandace_log.text
            f = open(f'HTMLs/{school}_{year}_gamelogs_advandce.html', 'w+')
            f.write(page_advandace_log)
            f.close()
            #print('advandce')
        except Exception as e:
            #print(e)
            pass
```
